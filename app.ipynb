```python
import cv2 as cv
from matplotlib import pyplot as plt

%matplotlib inline

image = cv.imread("me.jpeg")
image = cv.cvtColor(image, cv.COLOR_RGB2BGR)

print(image.shape)

plt.axis('off')
plt.imshow(image)
plt.show()
```

```python
import cv2 as cv
from matplotlib import pyplot as plt

%matplotlib inline

gray_image = cv.imread("me.jpeg", 0)

print(gray_image.shape)

plt.axis('off')
plt.imshow(gray_image, cmap="gray")
plt.show()
```

```python
import cv2 as cv
from matplotlib import pyplot as plt

%matplotlib inline

image = cv.imread("me.jpeg", 0)
thresh = 127
max = 255
ret, binary_image = cv.threshold(image, thresh, max, cv.THRESH_BINARY)

print(ret, binary_image)

plt.axis('off')
plt.imshow(binary_image, cmap="gray")
plt.show()
```

```python
import cv2 as cv
from matplotlib import pyplot as plt

%matplotlib inline

image = cv.imread("me.jpeg", 0)
max = 255
thresh = cv.adaptiveThreshold(image, max, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 21, 10)

print(thresh)

plt.axis("off")
plt.imshow(thresh, cmap="gray")
plt.show()

_, thresh = cv.threshold(image, max, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)

print(thresh)

plt.axis("off")
plt.imshow(thresh, cmap="gray")
plt.show()
```

```python
import cv2 as cv
from matplotlib import pyplot as plt

%matplotlib inline

image = cv.imread("me.jpeg", 0)
thresh = 127
max = 255
ret, binary_image = cv.threshold(image, thresh, max, cv.THRESH_BINARY)

print(ret, binary_image)

plt.axis('off')
plt.imshow(binary_image, cmap="gray")
plt.show()
```

```python
import cv2 as cv
from matplotlib import pyplot as plt

%matplotlib inline

image = cv.imread("me.jpeg", 0)
max = 255
thresh = cv.adaptiveThreshold(image, max, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 21, 10)

print(thresh)

plt.axis("off")
plt.imshow(thresh, cmap="gray")
plt.show()

_, thresh = cv.threshold(image, max, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)

print(thresh)

plt.axis("off")
plt.imshow(thresh, cmap="gray")
plt.show()
```

```python
import cv2 as cv
from matplotlib import pyplot as plt

%matplotlib inline

image = cv.imread("me.jpeg")
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
red = image[:, :, 0]
green = image[:, :, 1]
blue = image[:, :, 2]

for y in range(0, image.shape[0]):
    for x in range(0, image.shape[1]):
        if green[y][x] <= 220 or blue[y][x] >= 240:
            image[y][x] = 0

plt.axis("off")
plt.imshow(image, interpolation="none")
plt.show()
```

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline

image = cv.imread("me.jpeg", cv.IMREAD_GRAYSCALE)

plt.axis("off")
plt.imshow(image, cmap="gray")
plt.show()

mean = 50
sigma = 20
gaussian = np.random.normal(mean, sigma, (image.shape[0], image.shape[1]))
noise = image + gaussian

plt.axis("off")
plt.imshow(gaussian, cmap="gray")
plt.show()

plt.axis("off")
plt.imshow(noise, cmap="gray")
plt.show()

cv.imwrite("me_noise.jpeg", noise)

plt.axis("off")
plt.imshow(noise - image, cmap="gray")
plt.show()
```

```python
import cv2 as cv
from matplotlib import pyplot as plt

first_image = cv.imread("me.jpeg", cv.IMREAD_GRAYSCALE)
second_image = cv.imread("other_me.jpeg", cv.IMREAD_GRAYSCALE)
first_image = cv.resize(first_image, (second_image.shape[0], second_image.shape[1]))
alpha = 0.5
blended_image = cv.addWeighted(first_image, alpha, second_image, alpha, 0)

plt.axis("off")
plt.imshow(first_image, cmap="gray")
plt.show()

plt.axis("off")
plt.imshow(second_image, cmap="gray")
plt.show()

plt.axis("off")
plt.imshow(blended_image, cmap="gray")
plt.show()
```

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

sizes = (250, 250)
radius = sizes[0]

mask = np.zeros(sizes, dtype="uint8")
cv.circle(mask, sizes, radius, 255, -1)
plt.imshow(mask, cmap="gray")
plt.show()

mask = np.zeros(sizes, dtype="uint8")
cv.circle(mask, sizes, radius, 255, 1)
plt.imshow(mask, cmap="gray")
plt.show()
```

```python
import cv2 as cv
from matplotlib import pyplot as plt

%matplotlib inline

mask = np.zeros((960, 960), dtype="uint8")
cv.circle(mask, (600, 600), 400, 255, -1)

image = cv.imread("me.jpeg", cv.IMREAD_GRAYSCALE)

plt.axis("off")
plt.imshow(cv.bitwise_and(image, mask), cmap="gray")
plt.show()

plt.axis("off")
plt.imshow(cv.bitwise_xor(image, mask), cmap="gray")
plt.show()
```

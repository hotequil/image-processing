```python
import cv2 as cv
from matplotlib import pyplot as plt

%matplotlib inline

image = cv.imread("me.jpeg", 0)
thresh = 127
max = 255
ret, binary_image = cv.threshold(image, thresh, max, cv.THRESH_BINARY)

print(ret, binary_image)

plt.axis('off')
plt.imshow(binary_image, cmap="gray")
plt.show()
```

```python
import cv2 as cv
from matplotlib import pyplot as plt

%matplotlib inline

image = cv.imread("me.jpeg", 0)
max = 255
thresh = cv.adaptiveThreshold(image, max, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 21, 10)

print(thresh)

plt.axis("off")
plt.imshow(thresh, cmap="gray")
plt.show()

_, thresh = cv.threshold(image, max, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)

print(thresh)

plt.axis("off")
plt.imshow(thresh, cmap="gray")
plt.show()
```

```python
import cv2 as cv
from matplotlib import pyplot as plt

%matplotlib inline

image = cv.imread("me.jpeg", cv.IMREAD_GRAYSCALE)
image = cv.equalizeHist(image)

print(image)

plt.imshow(image * 2, cmap="Greys_r", vmin=0, vmax=255)
plt.show()
plt.hist(image.ravel(), 256, [0, 256])
plt.show()
```

```python
import cv2 as cv
from matplotlib import pyplot as plt

%matplotlib inline

image = cv.imread("me.jpeg")
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
red = image[:, :, 0]
green = image[:, :, 1]
blue = image[:, :, 2]

for y in range(0, image.shape[0]):
    for x in range(0, image.shape[1]):
        if green[y][x] <= 220 or blue[y][x] >= 240:
            image[y][x] = 0

plt.axis("off")
plt.imshow(image, interpolation="none")
plt.show()
```

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline

image = cv.imread("me.jpeg", cv.IMREAD_GRAYSCALE)

plt.axis("off")
plt.imshow(image, cmap="gray")
plt.show()

mean = 50
sigma = 20
gaussian = np.random.normal(mean, sigma, (image.shape[0], image.shape[1]))
noise = image + gaussian

plt.axis("off")
plt.imshow(gaussian, cmap="gray")
plt.show()

plt.axis("off")
plt.imshow(noise, cmap="gray")
plt.show()

cv.imwrite("me_noise.jpeg", noise)

plt.axis("off")
plt.imshow(noise - image, cmap="gray")
plt.show()
```
